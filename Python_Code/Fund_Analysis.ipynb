### Determining Good Fund using Random Foreset Classifier

### Predict Good fund for 3 years of investments

```python
# gather 3 year data for each fund
fund_data = pd.read_csv("../Fund_Data/Fund_Stats_Annualized_data/Fund_statistics_3years.csv").drop(u'Unnamed: 0',axis =1)
fund_data.head()
```

Expense ratio plays an imporant role while deciding fund's overall performance. Expense ratio is the price paid by investor. So we will consider net returns = returns - Expense Ratio.

```python
fund_data['Expense_Ratio'] = [0.0]*len(fund_data)
i=0
for ticker in fund_data.Fund_Ticker:
    try:
        fund_data['Expense_Ratio'][i] = fund_returns[ticker]['Annual_Net_Exp_Ratio'][-3:].mean()
        
    except:
        fund_data['Expense_Ratio'][i] = np.nan

    i=i+1

fund_data['Expense_Ratio'] = fund_data['Expense_Ratio'].ffill()
fund_data['Net_returns'] = fund_data['Return'] - fund_data['Expense_Ratio']

fund_data.head()
```

A good fund should give best returns and it should have the following values :

* R sqaured : between 85-100. 
* Return : as high as possible ( we consider return > 10 )
* Beta : greater than 1
* alpha : greater than 0
* sharpe ratio : greater than 0
* Standard deviation : low  ( we cosidered std deviation <15 )


```python
'''
convert quality to zero and 1
R_sqr = 85 and 100
beta = >1
alpha = >0
std < 9
Sharpe_Ratio>0
Net Returns>10
'''

def parse_returns(row):
    if((row.R_Squared >=85) & (row.R_Squared <=100) &
       (row.Beta > 1) &
       (row.Alpha >0) &
       (row.Std_Dev <15) &
       (row.Sharpe_Ratio  >0)&
       (row.Net_returns > 10)
       ):
        return 1
    else:
        return 0
```

Create a Y array  which gives 1 and 0 values for each fund based on parameters such as R squared value, alpha, beta, standard deviation, sharpe ratio, returns.

```python
# fund data and decides 1 or 0.
def create_score(fund_data):    
    Y = [0.0]*len(fund_data)
    for i in range(len(fund_data)):
        Y[i] = parse_returns(fund_data.ix[i])
    
    return Y

Y=create_score(fund_data)
```

Consider the alpha, beta, sharpe ratio, returns, standard deviation, rsquared parameters as X and good_or_bad indicator as Y

```python

X_df = fund_data.drop(fund_data.columns[[0,1,6,7,13]],1)
X= X_df.as_matrix()
Y = np.array(Y)
```

Use Random forest classifer and cross validation for number of trees ranging from 1 to 41 to find out which trees gives more accuracy. We will use f1 scoring parameter for cross validation

```python

# range of trees from 1 to 40
no_trees = range(1,41)

# number of Folds =10
k_fold=10 

# define a datafarme to store result of crosss validation for each iteration.
df_random_forest = pd.DataFrame()
df_random_forest['Num_Trees'] = [0]*len(no_trees)
df_random_forest['Score'] = [[]]*len(no_trees)

# compute score for various number of trees using RandomForestClassifier.
for i in no_trees:    
    forest = sklearn.ensemble.RandomForestClassifier(n_estimators=i)
    scores = sklearn.cross_validation.cross_val_score(forest, X, Y,scoring='f1', cv=k_fold)
    df_random_forest['Num_Trees'][i-1]= i
    df_random_forest['Score'][i-1]=scores

# Display head of result
df_random_forest.head()
```

```python
# plot the scores of the random forests as a function of the number of trees
 
plt.figure(figsize=(16,9))

# box Plot of Number of Trees Vs Score
sns.boxplot(df_random_forest.Score, names=df_random_forest.Num_Trees )

plt.title("Plot of Number of Trees Vs Score", fontsize=16)
plt.xlabel( "Number of Trees", fontsize=14)
plt.ylabel("Scores", fontsize=14)
plt.yticks(np.arange(0.3, 1.3, 0.1))
sns.set_context('poster')

```

Based on the above graph we can say that n_estimator =13 gives the best accuracy. So we will Build a random forest model with n_estimator = 13 for further analysis.


```python
## Train random forest classifier on the wine data using 13 trees. 
clf = sklearn.ensemble.RandomForestClassifier(n_estimators=13)
clf = clf.fit(X,Y)

# obtain the relative importance of the features 
feature_imp = clf.feature_importances_

#get column names
columns = X_df.columns

print feature_imp
```

```python
# Show a bar plot of feature imp of named features

index = np.arange(len(columns))
bar_width = 0.35
opacity = 0.4

plt.figure(figsize=(14,9))
plt.bar(index, feature_imp, bar_width,
                 alpha=opacity,
                 color='b',
                 label='')

plt.xlabel('Columns', fontsize =16)
plt.ylabel('Feature Importance', fontsize =16)
plt.title('Feature Importance for each column', fontsize =18)
plt.xticks(index, columns,rotation = 70)
plt.show()
```

Based on the above result we can say that Beta and Standard deviation and Return, R squared are the important fearures when determining the good mutual fund.

### Build Regression Model Based on Fund Returns data


Gather past 10 years fund returns for each fund and build a data frame containing past 10 years record for each fund as training set and current year data as test set.

```python
## Fit a regression model and Predict the results

X_train_all = pd.DataFrame()
Y_train_all = pd.DataFrame()
X_test_all = pd.DataFrame()
Y_test_all = pd.DataFrame()
ticker_arr = []
i=0
for ticker in fund_meta_data.Fund_Ticker:    # read returns for each fund
    
    try:
        
        fund_history=fund_returns[ticker]
        X = fund_history.drop(fund_history.columns[[0,1,8]],1).astype(float)     
        Y = fund_history[[fund_history.columns[1]]].astype(float)

        X.columns = [u'Risk', u'Category', u'+/- S&P', u'+/- Category', u'Annual_Net_Exp_Ratio', u'Turnover_Ratio']
        Y.columns = ['Returns']   

        # Training data contains past 10 years data
        X_train = X[:-1]
        Y_train = Y[:-1]

        #test data contains current years parameters
        X_test = X[-1:]  
        Y_test = Y[-1:]

        if((len(X_train)==0 )|(len(X_test) == 0 ) 
              | (len(Y_train) == 0)| (len(Y_test) == 0)
              | (len(X_train) != len(Y_train))):
            continue;
        
    except:     #ignore the funds for which data is not available
        continue;
    
    X_train_all = X_train_all.append(X_train)
    Y_train_all = Y_train_all.append(Y_train)
    
    
    X_test_all = X_test_all.append(X_test)
    Y_test_all =Y_test_all.append(Y_test)
    ticker_arr.insert(i,ticker)
    i=i+1
   
```

Build a regression model based on past 10 years data and predict returns for each fund
